```python
# 1. 必要なライブラリのインストールとデータのダウンロード
!pip install -U gensim
import nltk
nltk.download('brown')
from nltk.corpus import brown
import gensim

print("モデルの学習を開始します...（数十秒〜1分ほどかかります）")

# 2. word2vecモデルの学習 (skip-gram)
model = gensim.models.Word2Vec(brown.sents(), vector_size=100, seed=0, sg=1)

print("\n学習が完了しました！結果を出力します。\n" + "="*40)

# 3. 結果の確認
print("① 'mouth' のベクトル（先頭10次元のみ表示）:")
print(model.wv["mouth"][:10])

print("\n② 'mouth' に似た意味の単語ランキング:")
for word, sim in model.wv.most_similar('mouth', topn=10):
    print(f"  - {word}: {sim:.4f}")

print("\n③ 'computer' に似た意味の単語ランキング:")
for word, sim in model.wv.most_similar('computer', topn=10):
    print(f"  - {word}: {sim:.4f}")
```
